Splitting the dataset for training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop('Species',axis=1)
y=df['Species']
```

```python
X.head()
```

```python
y.head()
```

```python
y.tail()
```

Assigning 30% of the Dataset for Testing and remaining for Training

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

Using Logistic Regression to train and test the model

```python
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
lr_pred=lr.predict(X_test)
```

Importing Modules to get a clear picture on accuracy score and classification Report

```python
from sklearn.metrics import classification_report , accuracy_score
```

```python
print(accuracy_score(y_test,lr_pred))
```

```python
print(classification_report(y_test,lr_pred))
```

Let's Predict with User Input

```python
def classify_iris():
  sl=float(input("Enter Sepal Length : "))
  sw=float(input("Enter Sepal Width : "))
  pl=float(input("Enter Petal Length : "))
  pw=float(input("Enter Petal Width : "))
  user_data=pd.DataFrame([{'SepalLength':sl , 'SepalWidth':sw, 'PetalLength':pl, 'PetalWidth':pw}])
  prediction=lr.predict(user_data)
  print(f"The predicted Iris species is: {prediction[0]}")
classify_iris()
```
